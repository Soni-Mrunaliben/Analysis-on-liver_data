```python
import pandas as pd
import numpy as np
```

```python
# giving the variable name
liver_data1.columns=['Age','Gender','Total_Bilirubin','Direct_Bilirubin','Alkaline_Phosphotase',
'Alamine_Aminotransferase','Aspartate_Aminotransferase','Total_Protiens','Albumin',
'Albumin_and_Globulin_Ratio','Class']
```

```python
# So replacing the gender variable as Mode is male and it's a categorical variable
liver_data1["Gender"].fillna(liver_data1["Gender"].mode()[0], inplace=True)
```

```python
# Crosschkeing
liver_data1.isnull().sum()
```

```python
# As total protiens is a numerical variable so filling by median
liver_data1["Total_Protiens"].fillna(liver_data1['Total_Protiens'].median(), inplace = True)
```

```python
# As Albumin_and_Globulin_Ratio is a numerical variable so filling by median
liver_data1["Albumin_and_Globulin_Ratio"].fillna(liver_data1['Albumin_and_Globulin_Ratio'].median(), inplace = True)
```

```python
# Crosscheck
liver_data1.Albumin_and_Globulin_Ratio.isnull().sum()
```

```python
# crosscheck
liver_data1.isnull().sum()
```

```python
colname = ['Gender','Class'] # taking the obeject variable to perform label encoding
colname
```

```python
# Preprocessing the data
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for x in colname:
    liver_data1[x]  = le.fit_transform(liver_data1[x])
```

```python
# Crosscheck
liver_data1.head()
```

```python
# Create x and y
x = liver_data1.values[:,:-1] 
y = liver_data1.values[:,-1] # contain all rows and dependent variable
```

```python
 # scaling the data
# x and y are created as scaking is only perform on independent variables
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(x)

x = scaler.transform(x)
print(x)
```

```python
liver_data1.dtypes
```

```python
from sklearn.model_selection import train_test_split

#Spliting the data into test and train
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,
                                                    random_state=10)  
```

# Logistic Regression

```python
# Import 
from sklearn.linear_model import LogisticRegression

#Create the model
classifier = LogisticRegression()

#Fitting training data on the model
classifier.fit(x_train,y_train)

#Predicting on the xtest
y_pred = classifier.predict(x_test)
#print(list(zip(y_test,y_pred)))

print(classifier.coef_)
print(classifier.intercept_)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(y_test,y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(y_test,y_pred))

acc=accuracy_score(y_test, y_pred)
print("Accuracy of the model: ",acc)

```

```python
# Stored the predicted probabilities

y_pred_prob = classifier.predict_proba(x_test)
y_pred_prob
```

```python
for a in np.arange(0.4,0.61,0.01):
    predict_mine = np.where(y_pred_prob[:,1] > a, 1, 0)
    cfm=confusion_matrix(y_test, predict_mine)
    total_err=cfm[0,1]+cfm[1,0]
    print("Errors at threshold ", a, ":",total_err, " , type 2 error :", 
          cfm[1,0]," , type 1 error:", cfm[0,1])
```

```python
y_pred_class = []
for value in y_pred_prob[:,1]:
    if value > 0.57:
        y_pred_class.append(1)
    else:
        y_pred_class.append(0)
print(y_pred_class)
```

```python

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cfm=confusion_matrix(y_test,y_pred_class)
print(cfm)
acc=accuracy_score(y_test, y_pred_class)
print("Accuracy of the model: ",acc)
print(classification_report(y_test, y_pred_class))
```

# Upsampling

```python
liver_data1.Class.value_counts()
```

```python
from sklearn.utils import resample
# Separate majority and minority classes
df_majority = liver_data1[liver_data1.Class==0]
df_minority = liver_data1[liver_data1.Class==1]
 
# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=416,    # to match majority class
                                 random_state=10) # reproducible results
 
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
 
# Display new class counts
df_upsampled.Class.value_counts()
```

```python
X = df_upsampled.values[:,:-1]
Y = df_upsampled.values[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X)

X = scaler.transform(X)
print(X)
```

```python
from sklearn.model_selection import train_test_split

#Split the data into test and train
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,
                                                    random_state=10)  
```

```python
#predicting using the Decision_Tree_Classifier
from sklearn.tree import DecisionTreeClassifier

model_DecisionTree=DecisionTreeClassifier(random_state=10)

#fit the model on the data and predict the values
model_DecisionTree.fit(X_train,Y_train)
Y_pred=model_DecisionTree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,\
classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
## Conclusion(Decision Tree)
### It contains 2 data values and they are:
#### 0 mean it's not a patient and does not need diagnosed in future
#### 1 mean it's a patient and neww to be get diagnosed in future
### Here doing Decision Tree on training data and predicted the values on test data.
#### In this it has shown good overall accuracy(0.82) but the recall for predicting 1(it's a patient) is predicted 
#### good in this case which i.e.(0.93) even overall accuracy 1 is 0.83 which is very low 
### so, this algorithm is performing good in this dataset
```
